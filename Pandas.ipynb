```python
#group by operations
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi','Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj','Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd','B.Tech', 'MA', 'Msc', 'MA'],
         'Salary':[34234,12231,6564,3252,143424,32535,1000,45345]
        }
```

```python
type(data1)
```

```python
df_new=pd.DataFrame(data1)
df_new
```

```python
print(df_new.groupby('Name').groups)
```

```python
grp_name=df_new.groupby('Name')
grp_name
```

```python
#to print we can type case to list
list(grp_name)
```

```python
#or through iterattion
for i in grp_name:
    print(i)
```

```python
#grp by using multiple keys
df_new.groupby(['Name','Qualification'])
print(df_new.groupby(['Name','Qualification']).groups)
```

```python
df_new.groupby(['Address']).sum()
```

```python
#we can do like this also
df_new.groupby('Name').sum('Salary')
```

```python
#merging two dataframes
data1 = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'], 
        'Age':[27, 24, 22, 32], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']} 
   
# Define a dictionary containing employee data 
data2 = {'Name':['Abhi', 'Ayushi', 'Dhiraj', 'Hitesh'], 
        'Age':[17, 14, 12, 52], 
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj'], 
        'Qualification':['Btech', 'B.A', 'Bcom', 'B.hons']} 
```

```python
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
```

```python
df1
```

```python
df2
```

```python
#modigy the index
df1=pd.DataFrame(data1,index=[0,1,2,3])
df2=pd.DataFrame(data2,index=[4,5,6,7])
```

```python
df1
```

```python
df2
```
